```python
import cv2
import matplotlib.pyplot as plt
plt.set_cmap("gray")
im = cv2.imread("puppy.jpeg",0)              # Input image
plt.imshow(im)
plt.show()
d = {}                                       # Create empty dictionary
h,w = im.shape[0:2]
for i in range(h):
    for j in range(w):
        d[im[i,j]] = d.get(im[i,j],0)+1      # Increasing the count of every pixel                              

lst = []
for item in sorted(d):
    lst.append(d[item])                      # Create new list and append the sorted items

plt.plot(lst)
plt.show()

hist = cv2.calcHist([im],[0],None,[256],[0,256])    # In built function to calculate Histogram, used to compare our result
plt.plot(hist)
plt.show()
```
